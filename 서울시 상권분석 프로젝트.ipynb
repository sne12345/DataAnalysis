### 4.2 이상치 처리 - IQR 방식

```python
# IQR 기반 이상치 처리 함수
def outliers_iqr_upper(ys):
    quartile_1, quartile_3 = np.percentile(ys, [25, 75])
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    return upper_bound
def outliers_iqr_lower(ys):
    quartile_1, quartile_3 = np.percentile(ys, [25, 75])
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    return lower_bound
```

```python
# 상권 [추정매출] - [당월_매출_금액] 데이터 이상치 처리
upper_df7 = outliers_iqr_upper(df7['당월_매출_금액'])
lower_df7 = outliers_iqr_lower(df7['당월_매출_금액'])

df7 = df7[df7.당월_매출_금액 < upper_df7]
df7 = df7[df7.당월_매출_금액 > lower_df7]
```

```python
# 상권배후지 [추정매출] - [당월_매출_금액] 데이터 이상치 처리
upper_df16 = outliers_iqr_upper(df16['당월_매출_금액'])
lower_df16 = outliers_iqr_lower(df16['당월_매출_금액'])

df16 = df16[df16.당월_매출_금액 < upper_df16]
df16 = df16[df16.당월_매출_금액 > lower_df16]
```
